```python
import pandas as pd
```

```python
data = pd.read_csv('epidemiology1.csv')
# Tipo de dato de la fecha
data['date'] = pd.to_datetime(data['date'], format='%d/%m/%Y')

"""!!!!!!!!"""
# Rellenamos NA con 0
data['new_confirmed'].fillna(0, inplace=True)
data['new_recovered'].fillna(0, inplace=True)
data['new_deceased'].fillna(0, inplace=True)
"""!!!!!!!!"""

# Solo columnas necesarioas
cleaned_data = data[['date', 'new_confirmed', 'new_recovered', 'new_deceased']]
```
